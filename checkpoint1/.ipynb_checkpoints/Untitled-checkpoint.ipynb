```python
def remove_percents(df, col):
    df[col]= df[col].map(str)
    for ind in df.index:
#         df[col].apply(df[col][ind][:-1])
        df[col][ind] = df[col][ind][:-1]
#         df[col].str.slice(0,-1)
    return df


def fill_zero_iron(df):
    df['Iron (% DV)']= df['Iron (% DV)'].map(str)
    for ind in df.index:
        if df['Iron (% DV)'][ind] == 'na':
                df['Iron (% DV)'][ind] = 0
    return df

def fix_caffeine(df):
    return df


def fix_strings(df, col):
    df[col].str.lower()
    df[col].strings.str.replace('[^a-zA-Z]', '')

def standardize_names(df):
    return df






```
